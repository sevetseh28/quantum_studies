```python
from qiskit.visualization import plot_bloch_vector,plot_bloch_multivector
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Statevector

%matplotlib inline

qr = QuantumRegister(2, 'q')
qc = QuantumCircuit(qr)
qc.h(0)  # Apply Hadamard gate to qubit 0
qc.t(0)  # Apply T gate to qubit 0 - Same as a phase shift gate of /4

qc.h(1)
qc.tdg(1)  # Apply T-dagger gate to qubit 1 - Same as a phase shift gate of -/4

state = Statevector(qc)

state.draw('bloch')

# plot_bloch_multivector(state, title='Hadamard and T gates on two qubits')
```

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram

# Create a quantum circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)

# Apply Hadamard gate to put the qubit in superposition
qc.h(0)

# Measure the qubit into the classical bit
qc.measure(0, 0)

# Use Aer's QasmSimulator
simulator = Aer.get_backend('qasm_simulator')

# Execute the circuit on the simulator
job = execute(qc, simulator, shots=1024)

# Get the result
result = job.result()

# Get the counts (number of times each result occurred)
counts = result.get_counts(qc)

# Print the result
print("Counts:", counts)

# Plot the histogram
plot_histogram(counts, legend=["hello"], title='Hello THis is the TItle', figsize=(10, 5), color='green')
```

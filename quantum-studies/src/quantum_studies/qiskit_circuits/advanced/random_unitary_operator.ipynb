```python
from qiskit.quantum_info import random_hermitian, random_unitary
from qiskit import QuantumCircuit
from qiskit.providers.basicaer import BasicAer
from qiskit.visualization import plot_histogram, plot_state_qsphere
random_unitary_op = random_unitary(4) # Dimension 4 unitary matrix - 2 qubits


qc = QuantumCircuit(2)


qc.append(random_unitary_op, [0, 1])

# statevector_before_measurement = qc.snapshot(label="statevector_before_measurement")
qc_before_measurement = qc.copy()

qc.measure_all()

qc.draw("mpl")


backend = BasicAer.get_backend("qasm_simulator")
result = backend.run(qc).result()
counts = result.get_counts()

plot_histogram(counts)
```

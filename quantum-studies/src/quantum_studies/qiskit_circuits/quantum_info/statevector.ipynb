```python
from qiskit import QuantumCircuit

from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector

%matplotlib inline

qc = QuantumCircuit(2)
qc.h(0)  # Apply Hadamard gate to qubit 0
qc.cx(0, 1)  # Apply CNOT gate with qubit 0 as control and qubit 1 as target

statevector = Statevector(qc)

statevector.draw("latex")
```

```python

plot_bloch_multivector(statevector, title="Bell State |+ - Cannot be displayed as Bloch vectors")
# No arrows appear in the Bloch spheres for each qubit in an entangled 
# Bell state because the reduced states are maximally entangled, and thus have no well-defined Bloch vector to display.
```
